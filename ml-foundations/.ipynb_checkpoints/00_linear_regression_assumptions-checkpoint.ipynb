```python
import pandas as pd
```

# Load data

## Non-linear data

```python
df = pd.read_csv("airfoil_self_noise.dat", sep="\t", names=['freq', 'angle', 'chord', 'vel', 'thick', 'lvl'])
display(df)
```

```python
X_cols = ['freq', 'angle', 'chord', 'vel', 'thick']
y_cols = 'lvl'
X = df[X_cols]
y = df[y_cols]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y, y_pred)
```

## Linear data

```python
from sklearn.datasets import make_regression
```

```python
X_reg, y_reg = make_regression(n_samples=len(df), n_features=len(X_cols), noise=2, random_state=42)
```

```python
df_reg = pd.DataFrame()
for idx in range(X_reg.shape[1]):
    df_reg[f"X_{idx}"] = X_reg[:,idx]
df_reg["y"] = y_reg
```

```python
model = LinearRegression()
model.fit(X_reg, y_reg)
y_reg_pred = model.predict(X_reg)
```

```python
r2_score(y_reg, y_reg_pred)
```

## Normally distributed errors

```python
from statsmodels.stats.diagnostic import normal_ad
```

```python
diff = abs(y) - abs(y_pred)
```

```python
plt.hist(diff, bins=50)
plt.show()
```

```python
normal_ad(diff)[1]
```

```python
diff = abs(y_reg) - abs(y_reg_pred)
```

```python
plt.hist(diff, bins=50)
plt.show()
```

```python
normal_ad(diff)[1]
```

## Non autocorrelated errors

```python
from statsmodels.stats.stattools import durbin_watson
```

```python
diff = abs(y) - abs(y_pred)
```

```python
durbin_watson(diff)
```

```python
plt.scatter(y, diff)
plt.show()
```

```python
diff = abs(y_reg) - abs(y_reg_pred)
```

```python
durbin_watson(diff)
```

## Homoscedastic errors

```python
from statsmodels.stats.diagnostic import het_white
from scipy.stats import levene
```

```python
diff = abs(y) - abs(y_pred)
```

```python
levene(diff[:1000], diff[1000:])
```

```python
plt.scatter(x=range(len(diff)), y=diff, alpha=0.5)
```

```python
diff = abs(y_reg) - abs(y_reg_pred)
```

```python
levene(diff[:1000], diff[1000:])
```

```python
plt.scatter(x=range(len(diff)), y=diff, alpha=0.5)
```
